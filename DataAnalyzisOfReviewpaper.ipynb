```python
# Initialize a new column to store the comparison results
df['Similarity1'] = False

# Loop through each value in column A
for a in df['DOIwebofscience']:
    # Loop through each value in column B
    for b in df['DOIscopus']:
        # Check if the values are similar
        if a == b:
            df.loc[df['DOIwebofscience'] == a, 'Similarity'] = True
df.info()
df.head()   
```

```python
# Initialize a new column to store the comparison results
df['Similarity2'] = False

# Loop through each value in column A
for a in df['DOIscopus']:
    # Loop through each value in column B
    for b in df['DOIoldscopus']:
        # Check if the values are similar
        if a == b:
            df.loc[df['DOIscopus'] == a, 'Similarity'] = True
df.info()
df.head()   
```
